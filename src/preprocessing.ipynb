```python
import pandas as pd
from tqdm import tqdm

tqdm.pandas()

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
```

```python
df_invest = pd.read_excel('../data/raw/Data Scientist-Analyst position - Requirement.xlsx', sheet_name='Đầu tư', index_col='No')
df_invest.head(5)
```

```python
df_revenue = pd.read_excel('../data/raw/Data Scientist-Analyst position - Requirement.xlsx', sheet_name='Doanh thu', index_col='No')
df_revenue.head(5)
```

```python
def format_province(province):
    
    provinces = {'Đắk Lắk': ['Daklak'], 
                 'Khánh Hoà': ['Khánh Hòa'], 
                 'Thừa Thiên - Huế': ['TT Huế', 'Thừa Thiên Huế'], 
                 'Bà Rịa - Vũng Tàu': ['BRVT', 'BR-VT'], 
                 'Hồ Chí Minh': ['Tp Hồ Chí Minh']}
    
    for true_p, false_p in provinces.items():
        if province in false_p:
            province = true_p
    
    province = province.strip()

    words = province.split()

    titled_words = [word[0].upper() + word[1:].lower() for word in words]
    
    return ' '.join(titled_words)
```

```python
df_invest['Tỉnh'] = df_invest['Tỉnh'].progress_apply(format_province)
df_revenue['Tỉnh'] = df_revenue['Tỉnh'].progress_apply(format_province)
```

```python
df_invest.to_csv('../data/clean/invest_cleaned.csv', index='No', encoding='utf-8-sig')
df_revenue.to_csv('../data/clean/revenue_cleaned.csv', index='No', encoding='utf-8-sig')
```
